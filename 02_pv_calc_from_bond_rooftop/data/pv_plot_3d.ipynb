```python
#@title ## Завантажити дані сонячної фотоелектрогенерації, район Сигнівка, м. Львів
excel_file = '230717_building_footprints_Syhnivka_pv.xlsx'
sheet = 'Сегменти'
import pandas as pd
df = pd.read_excel(excel_file, sheet_name=sheet)
theta = df.slope # кут нахилу
phi = df.aspect # кут горизонтальної орієнтації
p = df.ppv_m2 # середньорічна потужність електрогенерації Вт/кв.м
```

plot_trisurf - це тип графіка поверхні з бібліотеки matplotlib, створеного шляхом тріангуляції компактних поверхонь скінченної кількості трикутників, які охоплюють всю поверхню таким чином, що кожна точка на поверхні є трикутником. Перетин будь-яких двох трикутників призводить до пустоти або спільного ребра чи вершини. Цей тип графіка створюється там, де сітки з рівномірною вибіркою є обмеженими та незручними для побудови. Зазвичай діаграми трьох поверхонь створюються шляхом виклику функції ax.plot_trisurf()  

```python
import matplotlib.pyplot as plt

fig = plt.figure()
axes = fig.add_subplot(projection='3d')
axes.scatter(theta, phi, p)

axes.set_xlim(0, 90)
axes.set_ylim(0, 360)

plt.show()
```
